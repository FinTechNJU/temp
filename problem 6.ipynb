### (a) Plot the PDF of the Gamma(3, 1) distribution

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import seaborn as sns

plt.rcParams['font.sans-serif'] = ['SimHei']  # 显示中文
sns.set_context("talk", font_scale=1)

# 设置画布
fig = plt.figure(figsize=(8, 8))  # 确定绘图区域尺寸
ax1 = fig.add_subplot(1, 1, 1)
x = np.arange(0.01, 20, 0.01) 

# 绘制gamma分布曲线
y3 = st.gamma.pdf(x, 3, scale=1)  # "α=3,β=2"


# 设置图例
ax1.plot(x, y3, label="α=3,β=2")

# 设置坐标轴标题
ax1.set_xlabel('x')
ax1.set_ylabel('p(x)')
ax1.set_title("PDF of the Gamma Distribution")
ax1.legend(loc="best")

plt.show()
```

### (c) Plot the parameters of the i-th posterior as well as the posterior mean and standard deviation.

The posterior distribution can be defined as:

$\pi(\theta|x)$ = $Gamma(\sum x_i + \alpha, n + \beta)$

the posterior mean      is $\frac{\sum x_i + \alpha}{n + \beta} $

the posterior deviation is $\frac{\sum x_i + \alpha}{(n + \beta)^2} $

where $\alpha = 3$ and $\beta = 1$ 


```python
import numpy as np 
lamb = 2
sample = np.random.poisson(lamb, size=500)
```

```python
x = np.arange(0,500)
```

```python
pmean = []
pdeviation = []
for i in range(500):
    pmean.append( ( sample[0:i+1].sum()+3 ) / (i+1 + 1) )
    pdeviation.append( ( sample[0:i+1].sum()+3 ) / (i+1 + 1)**2 )
```

Here is the parameter plot

```python
alpha = []
beta  = []
for i in range(500):
    alpha.append(  sample[0:i+1].sum()+3 )
    beta.append(  (i+1 + 1)  )
```

```python
plt.plot(x, alpha)
```

```python
plt.plot(x, beta)
```

We get the posterior mean and deviation: pmean and pdeviation

```python
plt.plot(x, pmean)
```

```python
plt.plot(x, pdeviation)
```

### (f) For each i from 1 to 500 compute and plot Bayes estimator for the quadratic loss function. 

### Also plot the 90%, 95%, and 99% confidence intervals around the Bayes estimator

The Bayes estimator is $\alpha / \beta $

```python
import numpy as np 
lamb = 2
sample = np.random.poisson(lamb, size=500)
falpha = lambda n: sample[0: n].sum()+3
fbeta  = lambda n: n + 1
listing = []
x = np.arange(500)
for i in range(500):
    listing.append(falpha(i)/fbeta(i))
```

Here is the plot of Bayes estimator

```python
plt.plot(x, listing)
```

For 95% prediction confidence, the interval is $[mean - 1.96 sigma, mean + 1.96 simga]$

The MLE of the parameter $\lambda$ is $\frac{1}{n}\sum x_i$

### (h) For each i from 1 to 500, plot MLE and standard error

```python
import numpy as np 
lamb = 2
sample = np.random.poisson(lamb, size=500)
listing = []
x = np.arange(500)
for i in range(500):
    listing.append(sample[0:i].sum())
```

```python
plt.plot(x, listing)
```
